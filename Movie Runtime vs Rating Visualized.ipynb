```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("complete_top_200_movies.csv")
df.head()
```

```python
#sorted_runtime = df.sort_values(by='Runtime', ascending = False)
# sorted_runtime['Runtime'].unique()
# Convert 'Runtime' column to strings
df['Runtime'] = df['Runtime'].astype(str)

df['Runtime'] = df['Runtime'].str.split().str[0].astype(int)
sorted_runtime = df.sort_values(by='Runtime')
sorted_runtime['Runtime'].unique()
```

```python
import numpy as np
runtime = sorted_runtime['Runtime']
rating = sorted_runtime['imdbRating']

# Scatter plot
plt.scatter(runtime, rating, color='gold', edgecolor='black', label='Movie')

# Linear regression
m, b = np.polyfit(runtime, rating, 1)
plt.plot(runtime, m*runtime + b, color='purple', label='Regression Line')

# Labels and title
plt.xlabel('Runtime')
plt.ylabel('IMDb Score (Out of 10)')
plt.title('Movie Runtime vs IMDb Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend()

plt.savefig('movie_runtime_vs_imdb_score.png')

plt.show()
```

```python
print(runtime.min())
print(runtime.max())
```

```python
money = sorted_runtime['Lifetime Gross']

# Scatter plot
plt.scatter(runtime, money, color='gold', edgecolor='black', label='Movie')

# Linear regression
m, b = np.polyfit(runtime, money, 1)
plt.plot(runtime, m*runtime + b, color='purple', label='Regression Line')

# Labels and title
plt.xlabel('Runtime')
plt.ylabel('Lifetime Grossing')
plt.title('Movie Runtime vs Lifetime Grossing')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend()

plt.savefig('movie_runtime_vs_lifetime_gross.png')

plt.show()
```
